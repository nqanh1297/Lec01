```python
#Nhập vào số nguyên dương n>2 hãy in ra ước số nguyên tố lớn nhất của n
import math
def snt(a):
    if a<2:
        return False
    for i in range (2,int(math.sqrt(a))+1):
        if(a%i==0): 
            return False
    return True
n=int(input())
uln=0
for i in range (2,n+1):
       if( n%i==0 and snt(i)):
           uln=i
print(uln)
```

```python
#nhập 2 số nguyên dương a,b hayz tính tổng số nguyên tố trong khoảng từ a đến b .mỗi dòng là một bộ test a b 
import math
def snt(a):
    if a<2:
        return False
    for i in range (2,int(math.sqrt(a))+1):
        if a%i==0:
            return False
    return True
n=int(input())
while(n!=0):
    n=n-1
    a=int(input())
    b=int(input())
    s=0
    for i in range(a,b+1):
        if(snt(i)): 
            s=s+i
    print(s)
```
