```python
#W5A1
def Max_Of_Two(a, b):
    if a == b:
       return "Hai số bằng nhau"
    elif a > b:
       return a 
    else:
       return b
a, b = map(int, input("Nhập hai số nguyên a, b: ").split())
print(Max_Of_Two(a, b))
    
```

```python
#W5A3
from math import *
def snt(n):
    for  i in range(2, int(sqrt(n)) + 1):
        if n % i == 0:
            return "Không là số nguyên tố"
    else:
        return "Là số nguyên tố"
n = int(input())
print(snt(n))
```

```python
#W5A4
def ppm(n):
    sum = 0
    for i in range(1, n):
        if n % i == 0:
            sum += i
    if sum == n:
        return "Là số hoàn hảo"
    else:
        return "Không là số hoàn hảo"
n = int(input())
print(ppm(n))
            
```

```python
#W5A5
def find(A, k):
    for i in range(len(A)):
        if A[i] == k:
            return i 
    else:
        return -1
A = list(map(int, input("Nhập list số nguyên: ").split()))
k = int(input("Nhập số nguyên k: "))
print(find(A, k))
```

```python
#W5A6
def gt(n):
    t = 1
    for i in range(1, n + 1):
        t *= i
    return t
n = int(input())
print(gt(n))
        
```

```python
#W5A7
def cclt(num1, num2, operat):
    if operat == "+":
        result = num1 + num2
    elif operat == "-":
        result = num1 - num2
    elif operat == "/":
        if num2 == 0:
            return "Không thể chia cho 0"
        result = num1 / num2
    elif operat == "*":
        result = num1 * num2
    else: 
        return "Vui lòng nhập dấu +, -, /, * "
    return f"{result:.2f}"
num1, num2 = map(float, input("Nhập 2 số bất kỳ: ").split())
operat = input("Nhập phép toán: ")
print(cclt(num1, num2, operat))
        
```

```python
#W5A8
def hamming(a, b):
    count = 0
    while a > 0 or b > 0:
        bita = a % 2
        bitb = b % 2
        if bita != bitb:
            count += 1
        a //= 2
        b //= 2   
    return count
a, b = map(int, input("Nhập hai số nguyên dương: ").split())
print(hamming(a, b))
    
    
```

```python
#W5A8
def hamming(a, b):
    count = 0
    x = a ^ b
    while x > 0:
        if x & 1 == 1:
            count +=1
        x >>= 1
    return count
a, b = map(int, input("Nhập hai số nguyên dương: ").split())
print(hamming(a, b))
```

```python
#W5A9
def s(n):
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total
n = int(input("Nhập số nguyên: "))
print(s(n))
        
```

```python
#W5A10
def check(a, b):
    og_a = [-1] * 256
    og_b = [-1] * 256
    if len(a) != len(b):
        return False
    for i in range(len(a)):
        x = a[i]
        y = b[i]
        indx = ord(x)
        indy = ord(y)
        if og_a[indx] != -1 and og_a[indx] != y:
            return False
        if og_b[indy] != -1 and og_b[indy] != x:
            return False
        og_a[indx] = y
        og_b[indy] = x
    return True
a, b = input("Nhập 2 dãy ký tự:").split()
print(check(a, b))
            

        
        
    
        
```
