```python
#W11A1
def reverse(path: str) -> int:
    try:
        with open(path, 'r') as file:
            data = file.read().strip()
            reversed_data = data[::-1]
            return int(reversed_data)
            
    except FileNotFoundError:
        print(f"Lỗi: Không tìm thấy tệp tại đường dẫn '{path}'")
        return -1
    except ValueError:
        print("Lỗi: Nội dung trong tệp không phải là số nguyên hợp lệ.")
        return -1
if __name__ == "__main__":
    filename = "data.txt"
    with open(filename, "w") as f:
        f.write("123456")
    print(f"Đã tạo file mẫu '{filename}' với nội dung: 123456")
    ket_qua = reverse(filename)
    print("Kết quả sau khi đảo ngược là:", ket_qua) 
```

```python
#W11A4
import os
def areAnagrams(path1: str, path2: str) -> bool:
    try:
        def clean(s):
            return sorted([c.lower() for c in s if c.isalnum()])

        with open(path1, 'r', encoding='utf-8') as f1:
            s1 = f1.read()
            
        with open(path2, 'r', encoding='utf-8') as f2:
            s2 = f2.read()

        return clean(s1) == clean(s2)
    except:
        return False

if __name__ == "__main__":
    file1 = "test1.txt"
    file2 = "test2.txt"
    
    with open(file1, "w") as f: f.write("Tom Marvolo Riddle")
    with open(file2, "w") as f: f.write("I Am Lord Voldemort")
    
    print(areAnagrams(file1, file2))
    
    os.remove(file1)
    os.remove(file2)
```

```python
#W11A8
def fibonacci_words(n):
    if n == 0:
        arr = ["a"]
    else:
        arr = ["a", "b"]
        for i in range(2, n + 1):
            next_str = arr[i-1] + arr[i-2]
            arr.append(next_str)
            
    with open("output.txt", "w") as f:
        for s in arr:
            print(s)
            f.write(s + "\n")

if __name__ == "__main__":
    n = 10
    fibonacci_words(n)
```

```python
#W11A10
import os

def getMoney(path: str) -> list[int]:
    try:
        with open(path, 'r') as f:
            content = f.read()
        return [int(x) for x in content.split()]
    except:
        return []

def rob(houses: list[int]) -> int:
    if not houses:
        return 0
    
    prev2 = 0
    prev1 = 0
    
    for money in houses:
        current_val = max(prev1, prev2 + money)
        prev2 = prev1
        prev1 = current_val
        
    return prev1

if __name__ == "__main__":
    filename = "money.txt"
    data = "1 2 3 1"
    
    with open(filename, "w") as f:
        f.write(data)
    
    houses_list = getMoney(filename)
    print(houses_list)
    
    max_money = rob(houses_list)
    print(max_money)
    
    os.remove(filename)
```

```python
#W11A12
import os

def maximumProduct(path: str) -> int:
    try:
        with open(path, 'r') as f:
            lines = f.readlines()
            if len(lines) < 2:
                return 0
            nums = [int(x) for x in lines[1].split()]
            
        nums.sort()
        
        opt1 = nums[-1] * nums[-2] * nums[-3]
        opt2 = nums[0] * nums[1] * nums[-1]
        
        return max(opt1, opt2)
    except:
        return 0

if __name__ == "__main__":
    filename = "test_max_product.txt"
    content = "5\n-10 -10 5 2 1"
    
    with open(filename, "w") as f:
        f.write(content)
        
    print(maximumProduct(filename))
    
    if os.path.exists(filename):
        os.remove(filename)
```

```python
#W11A13
import os

def moveZeroes(fileName: str) -> list:
    try:
        with open(fileName, 'r') as f:
            lines = f.readlines()
            if len(lines) < 2:
                return []
            nums = [int(x) for x in lines[1].split()]
            
        non_zeros = [x for x in nums if x != 0]
        zeros = [0] * (len(nums) - len(non_zeros))
        
        return non_zeros + zeros
    except:
        return []

if __name__ == "__main__":
    filename = "zeros_test.txt"
    content = "5\n0 1 0 3 12"
    
    with open(filename, "w") as f:
        f.write(content)
        
    print(moveZeroes(filename))
    
    if os.path.exists(filename):
        os.remove(filename)
```

```python
#W11A14
import os

def averageTime(path: str):
    try:
        with open(path, 'r') as f:
            lines = f.readlines()
        
        total_time = 0
        count = 0
        
        for line in lines[1:]:
            line = line.strip()
            if not line: continue
            
            parts = line.split(',')
            if len(parts) >= 2:
                total_time += float(parts[1])
                count += 1
                
        if count == 0:
            return 0.0
            
        return round(total_time / count, 2)
    except:
        return 0.0

if __name__ == "__main__":
    filename = "races.csv"
    content = "Lap,Time\n1,143.23\n2,134.12"
    
    with open(filename, "w") as f:
        f.write(content)
        
    print(averageTime(filename))
    
    if os.path.exists(filename):
        os.remove(filename)
```

```python
#W11A15
import os

def findMovies(path: str) -> list[str]:
    result = []
    try:
        with open(path, 'r') as f:
            lines = f.readlines()
            
        for line in lines[1:]:
            line = line.strip()
            if not line: continue
            
            parts = line.split(',')
            
            name = parts[1].strip()
            description = parts[2].strip()
            rating = float(parts[3].strip())
            
            if rating > 8.0 and description != "boring":
                result.append(name)
                
        if not result:
            return ["All are boring!"]
            
        return result
        
    except:
        return ["All are boring!"]

if __name__ == "__main__":
    filename = "movies.txt"
    content = """id, movie, description, rating
1, Thor, great 3D, 8.0
2, The Rings, Fantasy, 7.5
3, Irish, boring, 6.2
4, Ice song, Fantasy, 8.6
5, Home Alone, Interesting, 6.5"""
    
    with open(filename, "w") as f:
        f.write(content)
        
    movies = findMovies(filename)
    print(movies)
    
    if os.path.exists(filename):
        os.remove(filename)
```

```python
#W11A16
import math
import os

def calculate_perimeters(path: str):
    try:
        with open(path, 'r') as f:
            lines = [line.strip() for line in f if line.strip()]
        
        i = 0
        while i < len(lines):
            current_line = lines[i]
            
            if current_line.startswith("TYPE:"):
                shape_type = current_line.split(":")[1].strip()
                
                if i + 1 < len(lines):
                    param_line = lines[i+1]
                    parts = param_line.split()
                    
                    perimeter = 0
                    
                    if shape_type == "SQUARE":
                        side = float(parts[1])
                        perimeter = 4 * side
                        
                    elif shape_type == "RECTANGLE":
                        width = float(parts[1])
                        height = float(parts[3])
                        perimeter = 2 * (width + height)
                        
                    elif shape_type == "CIRCLE":
                        radius = float(parts[1])
                        perimeter = 2 * math.pi * radius
                        
                    print(f"{perimeter:.2f}")
                
                i += 2
            else:
                i += 1
                
    except:
        pass

if __name__ == "__main__":
    filename = "shapes.txt"
    content = """TYPE: SQUARE
SIDE: 5.3

TYPE: RECTANGLE
WIDTH: 5.0 HEIGHT: 16.0

TYPE: CIRCLE
RADIUS: 2.15"""
    
    with open(filename, "w") as f:
        f.write(content)
        
    calculate_perimeters(filename)
    
    if os.path.exists(filename):
        os.remove(filename)
```

```python
#W11A17
import os

def productExceptSelf(path: str) -> list[int]:
    try:
        with open(path, 'r') as f:
            content = f.read().split()
            
        if not content:
            return []
            
        # Bỏ qua số đầu tiên (n) vì python list tự quản lý độ dài
        # Lấy mảng bắt đầu từ phần tử thứ 2
        nums = [int(x) for x in content[1:]]
        
        n = len(nums)
        if n == 0:
            return []
            
        answer = [1] * n
        
        left = 1
        for i in range(n):
            answer[i] = left
            left *= nums[i]
            
        right = 1
        for i in range(n - 1, -1, -1):
            answer[i] *= right
            right *= nums[i]
            
        print(*answer)
        return answer
        
    except:
        return []

if __name__ == "__main__":
    filename = "test_case_17.txt"
    # Dữ liệu test case bạn cung cấp
    content = "5\n4 6 2 5 2"
    
    with open(filename, "w") as f:
        f.write(content)
        
    # Kết quả kỳ vọng: 120 80 240 96 240
    productExceptSelf(filename)
    
    if os.path.exists(filename):
        os.remove(filename)
```

```python
#W11A18
import math
import os

def estimatedTime(path: str):
    try:
        with open(path, 'r') as f:
            lines = f.readlines()

        src_parts = lines[0].split()
        x1 = float(src_parts[2])
        y1 = float(src_parts[4])

        des_parts = lines[1].split()
        x2 = float(des_parts[2])
        y2 = float(des_parts[4])

        vel_parts = lines[2].split()
        velocity = float(vel_parts[1])

        distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
        
        if velocity == 0:
            return 0.0

        return round(distance / velocity, 2)
    except:
        return 0.0

if __name__ == "__main__":
    filename = "rocket_config.txt"
    content = """SRC X: 0 Y: 0
DES X: 10000 Y: 10000
VELOCITY: 6 KM/S"""
    
    with open(filename, "w") as f:
        f.write(content)
        
    print(estimatedTime(filename))
    
    if os.path.exists(filename):
        os.remove(filename)
```

```python
#W11A19
import struct
import os

def read_binary_doubles(path: str):
    try:
        with open(path, 'rb') as f:
            while True:
                chunk = f.read(8)
                if len(chunk) != 8:
                    break
                
                number = struct.unpack('d', chunk)[0]
                print(f"{number:.5f}")
    except:
        pass

if __name__ == "__main__":
    # Phần tạo file giả lập để test (do không có sẵn file nhị phân)
    filename = "test_numbers.bin"
    values = [3.1415926535, 10.5, -0.001234, 123.456789]
    
    with open(filename, 'wb') as f:
        for val in values:
            f.write(struct.pack('d', val))

    # Trong bài làm thực tế, bạn có thể dùng: path = input()
    read_binary_doubles(filename)
    
    if os.path.exists(filename):
        os.remove(filename)
```

```python
#W11A20
import os

def imageSmoother(path: str):
    try:
        with open(path, 'r') as f:
            content = f.read().split()
            
        if not content:
            return

        M = int(content[0])
        raw_data = [int(x) for x in content[1:]]
        
        img = []
        for i in range(M):
            start = i * M
            end = start + M
            img.append(raw_data[start:end])
            
        result = [[0] * M for _ in range(M)]
        
        for r in range(M):
            for c in range(M):
                total = 0
                count = 0
                
                for i in range(r - 1, r + 2):
                    for j in range(c - 1, c + 2):
                        if 0 <= i < M and 0 <= j < M:
                            total += img[i][j]
                            count += 1
                            
                result[r][c] = total // count
                
        for row in result:
            print(*(row))
            
    except:
        pass

if __name__ == "__main__":
    filename = "matrix_test.txt"
    content = """4
130 160 150 170
160 140 180 160
170 190 120 150
210 180 160 120"""
    
    with open(filename, "w") as f:
        f.write(content)
        
    imageSmoother(filename)
    
    if os.path.exists(filename):
        os.remove(filename)
```
